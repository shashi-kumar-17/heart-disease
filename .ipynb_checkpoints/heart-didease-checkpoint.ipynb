# Predicting Heart Disease using Machine Learning

### Objective
##### The goal is to build a machine learning model that predicts disease outbreaks or assesses the risk of certain health conditions based on historical data. This aligns with the broader objective of supporting early detection, prevention, and effective management of health issues, contributing to [`SDG 3`](https://jointsdgfund.org/sustainable-development-goals/goal-3-good-health-and-well-being) (Good Health and Well-being).

### Data

The original data came from the [Cleveland database](https://archive.ics.uci.edu/dataset/45/heart+disease) from UCI Machine Learning Repository.

Howevever, we've downloaded it in a formatted way from [Kaggle](https://www.kaggle.com/datasets/sumaiyatasmeem/heart-disease-classification-dataset).

The original database contains 76 attributes, but here only 14 attributes will be used.

### Creating Model

```python
X = df.drop("target", axis=1)

y = df["target"]
```

```python
X.head(3)
```

```python
y.head(3)
```

```python
np.random.seed(42)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
models = {"Logistic Regression": LogisticRegression(),
          "KNN": KNeighborsClassifier(),
          "Random Forest": RandomForestClassifier()}

def fit_and_score(models, X_train, X_test, y_train, y_test):
    """
    Fits and evaluates given machine learning models.
    models : a dict of differetn Scikit-Learn machine learning models
    X_train : training data (no labels)
    X_test : testing data (no labels)
    y_train : training labels
    y_test : test labels
    """
    np.random.seed(42)
    model_scores = {}
    for name, model in models.items():
        model.fit(X_train, y_train)
        model_scores[name] = model.score(X_test, y_test)
    return model_scores
```

```python
model_scores = fit_and_score(models=models,
                             X_train=X_train,
                             X_test=X_test,
                             y_train=y_train,
                             y_test=y_test)
```

```python
model_scores
```

```python
model_compare = pd.DataFrame(model_scores, index=["accuracy"])
model_compare.T.plot.bar()
```

### Conclusion
    Hence, the Logistic Reression model gives more accurate values compared to KNN and Random Forest Classifiers for this dataset.
